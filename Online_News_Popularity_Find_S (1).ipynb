```python
import pandas as pd

data = pd.DataFrame({
    "Category": ["Technology", "Technology", "Technology", "Sports", "Sports",
                 "Entertainment", "Technology", "Entertainment", "Sports", "Technology"],
    "Platform": ["Social", "Social", "Traditional", "Traditional", "Social",
                 "Social", "Social", "Traditional", "Traditional", "Social"],
    "Publish_Day": ["Weekday", "Weekend", "Weekday", "Weekend", "Weekday",
                    "Weekday", "Weekday", "Weekend", "Weekday", "Weekend"],
    "Article_Size": ["Short", "Short", "Short", "Long", "Short",
                     "Long", "Long", "Short", "Long", "Short"],
    "Engagement": ["High", "High", "High", "Low", "Low",
                   "Low", "High", "Low", "Low", "High"]
})

data
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

print("X =", X)
print("y =", y)
```

```python
num_attr = X.shape[1]

S = ['Ø'] * num_attr
G = [['?'] * num_attr]

print("Initial S =", S)
print("Initial G =", G)
```

```python
for i in range(len(X)):
    if y[i] == "High":   # Positive example
        for j in range(num_attr):
            if S[j] == 'Ø':
                S[j] = X[i][j]
            elif S[j] != X[i][j]:
                S[j] = '?'

        G = [g for g in G if all(g[j] == '?' or g[j] == S[j] for j in range(num_attr))]

    else:  # Negative example
        new_G = []
        for g in G:
            for j in range(num_attr):
                if g[j] == '?' and S[j] != X[i][j]:
                    temp = g.copy()
                    temp[j] = S[j]
                    new_G.append(temp)
        G = new_G

    print(f"\nAfter example {i+1}")
    print("S =", S)
    print("G =", G)
```

```python
# Remove duplicate hypotheses from G
G = [list(h) for h in set(tuple(g) for g in G)]

print("\nFinal Specific Hypothesis (S):")
print(S)

print("\nFinal General Hypothesis (G):")
print(G)
```
